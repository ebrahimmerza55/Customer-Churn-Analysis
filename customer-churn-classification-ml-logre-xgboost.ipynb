```python
def replace_yes_no(df, columns):
    
    for col in columns:
        df[col] = df[col].replace({'Yes': 1, 'No': 0})

columns_to_replace = [
    "Partner", "Dependents", "PhoneService", "OnlineSecurity", "OnlineBackup", 
    "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", 
    "PaperlessBilling", "Churn"
]

replace_yes_no(df, columns_to_replace)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
columns = ["MultipleLines", "Contract", "PaymentMethod", "InternetService"]

for column in columns:
    df[column] = labelencoder.fit_transform(df[column])
    
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['TotalCharges'].fillna(value=df['TotalCharges'],inplace=True)
```

```python
df = df.dropna()
df.isna().sum()
```

## Build Models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
X = df.drop(columns=['Churn'])
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## LogisticRegression

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test_scaled)
```

```python
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)

print(f'Logistic Regression Accuracy: {accuracy_log_reg:.2f}')
```

```python
disp_log_reg = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_log_reg, display_labels=['No Churn', 'Churn'])
disp_log_reg.plot(cmap='Blues')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

## RandomForest 

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
y_pred_rf = rf.predict(X_test_scaled)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

print(f'Random Forest Accuracy: {accuracy_rf:.2f}')
```

```python
disp_rf = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_rf, display_labels=['No Churn', 'Churn'])
disp_rf.plot(cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

## FineTuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6, 9],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
```

```python
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, scoring='accuracy', cv=5, verbose=1)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
best_xgb = grid_search.best_estimator_
y_pred_xgb = best_xgb.predict(X_test_scaled)
```

```python
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)

print(f'XGBoost Accuracy after tuning: {accuracy_xgb:.2f}')
```
